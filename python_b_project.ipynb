Use pandas to read in data file that has number of community residential facilities and suicide rates in adolescents in north american countries.

Variables were created for number of facilities and suicide rates.

Function was created to make scatterplot using the two data sets.

Linear regression was also conducted using scipy stats libraries.

We also created a function to make a bar graph using the two sets of data.

```python
data = pd.read_csv("/content/drive/MyDrive/python_b_data.csv")

health_fac = data['community residential facilities (per 100,000 population)']
suicide_rates = data['suicide rates (per 100,000 population)']

def scatterplot():
  plt.style.use('seaborn')
  plt.scatter(health_fac, suicide_rates, s=75, alpha=0.8, edgecolor='black', linewidth=1)

  plt.title('Suicide Rates vs Number of Community Residential Facilities in North American Countries')
  plt.xlabel('Community Residential Mental Health Facilities (per 100,000 people)')
  plt.ylabel('Suicide Rates (per 100,000 people)')
  sb.regplot(x=health_fac, y=suicide_rates, color='mediumpurple')

  plt.tight_layout()
  plt.show()

scatterplot()

health_fac = [0,0,0.704,0.278,0,0.042,1.213,0,0,0,0.006,0,0.011,2.994,0.012,0.033,0,0,0,0,0]
suicide_rates = [0,1.7,0,3.6,8.6,8.9,3.7,2.7,11.3,0,10.8,8.1,1.3,1.1,6.6,6.4,3.3,2.4,0,8.3,9.4]

result = scipy.stats.linregress(health_fac, suicide_rates)
slope = result.slope
rvalue = result.rvalue
print("Slope: " + str(slope))
print("R-value: " + str(rvalue))
corr, _ = pearsonr(health_fac, suicide_rates)
print('Pearsons correlation coefficient: %.3f' % corr)

def bargraph():
  bargraph_data = pd.read_csv("/content/drive/MyDrive/python_b_bargraph_data2.csv")
  df = pd.DataFrame(bargraph_data)
  X = list(df.iloc[:, 0])
  Y = list(df.iloc[:, 1])
  plt.bar(X, Y, color='lightpink')
  plt.title("Average Suicide Rate vs Number of Community Residential Facilities in North American Countries", fontsize=10)
  plt.xlabel("Number of Community Residential Facilities")
  plt.ylabel("Average Suicide Rates (per 100,000 people)")
  plt.show()

bargraph()
```
