```python
data = pd.read_csv("/content/drive/MyDrive/python_b_data.csv")

health_fac = data['community residential facilities (per 100,000 population)']
suicide_rates = data['suicide rates (per 100,000 population)']

plt.style.use('seaborn')
plt.scatter(health_fac, suicide_rates, s=75, alpha=0.8, edgecolor='black', linewidth=1)

plt.title('Suicide Rates vs Number of Community Residential Facilities in North American Countries')
plt.xlabel('Community Residential Mental Health Facilities (per 100,000 people)')
plt.ylabel('Suicide Rates (per 100,000 people)')
sb.regplot(x=health_fac, y=suicide_rates)

plt.tight_layout()
plt.show()

countries = ['AG','BAM','BRDO','BLZ','CA','CR','CU','DO','ELSL','GD','GT','HT','HN','JAM','MX','NIC','PA','LC','VC','TT','USA']
health_fac = [0,0,0.704,0.278,0,0.042,1.213,0,0,0,0.006,0,0.011,2.994,0.012,0.033,0,0,0,0,0]
suicide_rates = [0,1.7,0,3.6,8.6,8.9,3.7,2.7,11.3,0,10.8,8.1,1.3,1.1,6.6,6.4,3.3,2.4,0,8.3,9.4]

result = scipy.stats.linregress(health_fac, suicide_rates)
#print(result)
slope = result.slope
rvalue = result.rvalue
print("Slope: " + str(slope))
print("R-value: " + str(rvalue))
corr, _ = pearsonr(health_fac, suicide_rates)
print('Pearsons correlation coefficient: %.3f' % corr)


x = np.arange(len(countries)) # the label locations
width = 0.50 # the width of the bars

fig, ax = plt.subplots()

ax.set_ylabel('Suicide Rates (per 100,000 people')
ax.set_xlabel("Number of Community Residential Facilities (per 100,000 people)")
ax.set_title('Suicide Rates vs Number of Community Residential Facilities in North American Countries', fontsize=10)
ax.set_xticks(x)
ax.set_xticklabels(countries, fontsize=6)

pps = ax.bar(x - width/2, suicide_rates, width, label='suicide rates')
for p in pps:
   height = p.get_height()
   ax.annotate('{}'.format(height),
      xy=(p.get_x() + p.get_width() / 2, height),
      xytext=(0, 2), # 3 points vertical offset
      textcoords="offset points",
      ha='center', va='bottom')

plt.show()


```
